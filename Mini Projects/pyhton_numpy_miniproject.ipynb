The objective of this mini project is to analyze student performance using the NumPy library.
This project focuses on performing efficient numerical operations such as:
-Handling missing values
-Calculating averages
-Assigning grades
-Ranking students
-using NumPy arrays.

```python
import numpy as np
import sys
```

```python
print("===== STUDENT PERFORMANCE ANALYSIS SYSTEM (NumPy) =====\n")
```

```python
#Studnets names
students = np.array(["Ravi",
                     'Arjun',
                     "Ram",
                    "Mahesh",
                    "Kalayan"])
#Subjects
subjects = np.array(["maths", "English", "Science"])
#corresponding marks
marks = np.array([
    [85, 78, 92],
    [88, np.nan, 79],
    [72, 68, 70],
    [90, 95, np.nan],
    [60, 65, 58]
])
```

Handling missing values

```python
subject_means = np.nanmean(marks, axis=0)
for i in range(marks.shape[1]):
    marks[np.isnan(marks[:, i]), i] = subject_means[i]

print("Marks after handling missing values:\n", marks, "\n")
```

```python
#studnets wise  average
std_avg = np.mean(marks, axis = 1)
```

```python
#subject wise avergae
subject_avg = np.mean(marks, axis=0)
print("Subject-wise Average:")
for i in range(len(subjects)):
    print(subjects[i], ":", round(subject_avg[i], 2))
print()
```

```python
#Grades assignment
grades = []
for avg in std_avg:
    if avg >= 85:
        grades.append("A")
    elif avg >= 70:
        grades.append("B")
    elif avg >= 60:
        grades.append("C")
    else:
        grades.append("Fail")
grades = np.array(grades)
```

```python
#Ranking the studnets
ranks = np.argsort(-std_avg) + 1
```

```python
#Dispalying Final Studnets report
print("FINAL STUDENT REPORT")
print("----------------------")
for i in range(len(students)):
    print(
        f"Name: {students[i]:6} | "
        f"Average: {std_avg[i]:.2f} | "
        f"Grade: {grades[i]} | "
        f"Rank: {ranks[i]}"
    )
```

```python
#Finding top student and weak students
top_student = students[np.argmax(std_avg)]
weak_students = students[std_avg < 60]
print("\nTop Performer:", top_student)

if len(weak_students) > 0:
    print("Weak Students:", weak_students)
else:
    print("No weak students")
```
