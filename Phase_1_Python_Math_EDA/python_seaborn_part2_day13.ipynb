#Histograms in Seaborn

Histograms are visualization tools that represent the distribution of a set of continuous data. In a
histogram, the data is divided into a set of intervals or bins (usually on the x-axis) and the count of 
data points that fall into each bin corresponding to the height of the bar above that bin

Syntax: seaborn.histplot(data, x, y, hue, stat, bins, binwidth, discrete, kde, log_scale)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data = penguins, x = "body_mass_g", kde = True)
```

```python
sns.histplot(data = penguins, x = "body_mass_g", kde = True, hue = "species")
```

```python
tips = sns.load_dataset("tips")
sns.histplot(data = tips, x = "size", stat = "probability", discrete = True)
```

#join plot in Seabron

This function provides a convenient interface to the 'JointGrid' class, with several canned plot kinds

Syntax: seaborn.jointplot(x,  y,  data=None, kind='scatter', stat_func=None, color=None, height=6, ratio=5, space=0.2, dropna=True,  xlim=None, ylim=None, joint_kws=None, marginal_kws=None, annot_kws=None, **kwargs)

```python
data = sns.load_dataset("attention")
sns.jointplot(x = "solutions", y = "score",
              kind = "hex", data = data)
plt.show()
```

```python
data = sns.load_dataset("mpg")
sns.jointplot(x = "mpg", y = "acceleration",
              kind = "scatter", data = data)
plt.show()
```

#Pairplot in Seaborn

seaborn.pairplot() method is used for visualizing relationships between multiple variables in a dataset

```python
df = sns.load_dataset('tips')
custom_palette = {'Male': 'lightblue', 'Female': 'pink'}
sns.pairplot(df, hue='sex', palette=custom_palette)
plt.show()
```

#KDE plot in Seaborn

Kernel Density Estimate (KDE) Plot is a powerful tool for estimating the probability density function of continuous or non-parametric data

```python
import numpy as np
%matplotlib inline
```

```python
x= np.random.randn(200)
y = np.random.randn(200)
sns.kdeplot(x)
```

```python
sns.kdeplot(y)
```

```python
sns.kdeplot(x, shade = True)
```

```python
sns.kdeplot(x, shade = True, color='Green')
```

#Heatmap in Seaborn

A heatmap is a graphical representation of data where individual values are represented by color 
intensity. It is widely used in data analysis and visualization to identify patterns, correlations and 
trends within a datase

Syntax - 
seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, annot_kws=None, linewidths=0, 
linecolor='white', cbar=True, **kwargs)

```python
data = np.random.randint(1, 100, (10, 10))
sns.heatmap(data, vmin=30, vmax=70)
```

```python
data = np.random.randint(1, 100, (10, 10))
sns.heatmap(data, cmap='coolwarm', center=50)
plt.show()
```

#lmplot() in Seaborn

seaborn.lmplot() method is used to draw a scatter plot onto a FacetGrid.

Syntax : seaborn.lmplot(x, y, data, hue=None, col=None, row=None, palette=None, col_wrap=None, height=5, aspect=1, markers='o', sharex=True, sharey=True, hue_order=None, col_order=None, row_order=None,  legend=True, legend_out=True, x_estimator=None, x_bins=None, x_ci='ci', scatter=True, fit_reg=True, ci=95, n_boot=1000, units=None, seed=None, order=1, logistic=False, lowest=False, robust=False,logx=False, x_partial=None, y_partial=None, truncate=True, x_jitter=None, y_jitter=None,scatter_kws=None, line_kws=None, size=None)

```python
import pandas as pd
df = sns.load_dataset('tips')
sns.lmplot(x ='total_bill', y ='tip', data = df)
plt.show()
```

```python
sns.lmplot(x ='total_bill', y ='tip',
           fit_reg = False, data = df)
plt.show()
```

#regplot() in Seaborn

This method is used to plot data and a linear regression model fit. There are a number of mutually
exclusive options for estimating the regression model. For more information

Syntax : 
seaborn.regplot( x,  y,  data=None, x_estimator=None, x_bins=None,  x_ci='ci', scatter=True, fit_reg=True, ci=95, n_boot=1000, units=None, order=1, logistic=False, lowess=False, robust=False, logx=False,  x_partial=None, y_partial=None, truncate=False, dropna=True, x_jitter=None, y_jitter=None, label=None, color=None, marker='o',    scatter_kws=None, line_kws=None, ax=None)

```python
data = sns.load_dataset("mpg")
sns.regplot(x = "mpg", 
            y = "acceleration", 
            data = data)
plt.show()
```

```python
data = sns.load_dataset("titanic")
sns.regplot(x = "age",
            y = "fare",
            data = data,
            dropna = True)
plt.show()
```
