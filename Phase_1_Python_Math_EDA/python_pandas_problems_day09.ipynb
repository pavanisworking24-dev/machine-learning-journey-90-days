#Problem 5
Problem Statement: Creating Pandas DataFrame
Write a Python program that reads Name, Age, and Roll no. as input from the
user and stores it into a DataFrame, and printsthe DataFrame you have created.

```python
n = int(input("Enter number of students: "))

names = []
ages = []
roll_nos = []

for i in range(n):
    print(f"\nEnter details for student {i+1}")
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    roll_no = int(input("Enter Roll No: "))

    names.append(name)
    ages.append(age)
    roll_nos.append(roll_no)

# Create DataFrame
data = {
    "Name": names,
    "Age": ages,
    "Roll_no": roll_nos
}

df = pd.DataFrame(data)
print("\nStudent DataFrame:")
print(df)
```
