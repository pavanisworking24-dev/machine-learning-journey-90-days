#Loops in Python - For, While and Nested Loops

```python
#for loop - is used to iterate over a sequence such as a list, tuple, string or range
n = 7
for i in range(n):
    print(i)
```

```python
#list
lst = [1, 2, 3]
for x in lst:
    print(x)
#tuple
t1 = (45, 3, 89)
for x in t1:
    print(t1)
#set
s1 = {3, 67, 12}
for x in s1:
    print(x)
```

```python
#while loop
cnt = 0
while (cnt < 3):
    cnt = cnt + 1
    print("Hello Geek")
```

```python
#nested loops
for i in range(6):
    for j in range(i):
        print(i, end=" ")
    print()
```
