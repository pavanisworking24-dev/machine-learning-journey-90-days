#Data Normalization with Pandas

Data normalization means scaling numeric features to a similar range so that no feature dominates 
others just because of its size.

Avoids numerical instability in models
Speeds up convergence in gradient-based algorithms
Ensures all features contribute equally to the analysis

```python
import pandas as pd  
import matplotlib.pyplot as plt  

df = pd.DataFrame([
    [180000, 110, 18.9, 1400],  
    [360000, 905, 23.4, 1800],  
    [230000, 230, 14.0, 1300],  
    [60000, 450, 13.5, 1500]
], columns=['Col A', 'Col B', 'Col C', 'Col D'])
df
```

```python
df.plot(kind='bar')
plt.show()
```

```python
#Normalization Techniques in Pandas
```

#Min Max scaling
This technique rescales each feature between -1 and 1 by dividing all values by the maximum absolute
value in that column. 

```python
max_scaled = df.copy()
for column in max_scaled.columns:
    max_scaled[column] = max_scaled[column] / max_scaled[column].abs().max()
print(max_scaled)
```

```python
max_scaled.plot(kind = 'bar')
plt.show()
```

#The min-max feature scaling

The min-max approach also called normalization rescales the feature to a hard and fast range of
[0,1] by subtracting the minimum value of the feature then dividing by the range.

```python
scaled = df.copy()

for column in scaled.columns:
    scaled[column] = (scaled[column] - scaled[column].min()) / (scaled[column].max() - scaled[column].min())

print(scaled)
```

```python
scaled.plot(kind='bar')
plt.show()
```

#The z-score method

The z-score method often called standardization changes the values in each column so that they have a
mean of 0 and a standard deviation of 1.

```python
z_scaled = df.copy()

for column in z_scaled.columns:
    z_scaled[column] = (z_scaled[column] - z_scaled[column].mean()) / z_scaled[column].std()

print(z_scaled)

z_scaled.plot(kind='bar')
plt.show()
```

#Groupby() method in Pandas Dataframe

Pandas groupby() function is a powerful tool used to split a DataFrame into groups based on one or 
more columns, allowing for efficient data analysis and aggregation

```python
df_data = pd.read_csv("nba.csv")
df_data
```

```python
#grouping by single column
team = df_data.groupby('Team')
team.first().head(7)
```

```python
team = df_data.groupby(['Team', 'Position'])
team.first()
```

```python
# Applying Aggregation with GroupBy
aggregated_data = df_data.groupby(['Team', 'Position']).agg(
    total_salary=('Salary', 'sum'),
    avg_salary=('Salary', 'mean'),
    player_count=('Name', 'count')
)

print(aggregated_data)
```

```python
# Rank players within each team by Salary
df_data['Rank within Team'] = df_data.groupby('Team')['Salary'].transform(lambda x: x.rank(ascending=False))
print(df_data)
```
