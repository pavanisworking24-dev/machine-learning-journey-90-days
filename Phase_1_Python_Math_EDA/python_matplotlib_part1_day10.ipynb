#Error graphs in matplotlib in pyhton
Error bars are a graphical overlay used to display the variability or uncertainty of points
plotted on a Cartesian graph.
- Small error bar â†’ results are stable & reliable 
- Large error bar â†’ results are noisy & uncertain

```python
#Basic Graph
x =[1, 2, 3, 4, 5, 6, 7]
y =[1, 2, 1, 2, 1, 2, 1]
plt.plot(x, y)
```

```python
#error along x-axis
x =[1, 2, 3, 4, 5, 6, 7]
y =[1, 2, 1, 2, 1, 2, 1]
x_error = 0.5
plt.plot(x, y)
plt.errorbar(x, y, xerr = x_error, fmt='o')
```

```python
#error along y-axis
x =[1, 2, 3, 4, 5, 6, 7]
y =[1, 2, 1, 2, 1, 2, 1]
y_error = 0.5
plt.plot(x, y)
plt.errorbar(x, y, yerr = y_error, fmt='o')
```

```python
#error along in both x and y-axis
x =[1, 2, 3, 4, 5, 6, 7]
y =[1, 2, 1, 2, 1, 2, 1]
x_error = 0.5
y_error = 0.5
plt.plot(x, y)
plt.errorbar(x, y, yerr = y_error, xerr = x_error, fmt='o')
```

#Line chart in Matplotlib 

Line charts are used to represent the relation between two data X and Y on a different axis.

```python
x = np.array([1, 2, 3, 4])
y = x*2
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("line chart")
plt.show()
```

```python
#Adding annotations
x = np.array([1, 2, 3, 4])
y = x*2
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords = "offset points", xytext = (0, 10), ha = 'center')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("line chart")
plt.grid(True)
plt.show()
```

```python
#Multiple Line Charts Using Matplotlib 
x = np.array([1, 2, 3, 4])
y = x*2
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Any suitable title")
plt.show()  # show first chart

plt.figure()
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
plt.plot(x1, y1, '-.')
plt.show()
```

```python
#Multiple Plots on the Same Axis
x = np.array([1, 2, 3, 4])
y = x*2
# first plot with X and Y data
plt.plot(x, y)
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.show()
```

#Bar Plot in Matplotlib

A bar plot uses rectangular bars to represent data categories, with bar length or height proportional 
to their values. It compares discrete categories

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#bar chart
fruits = ['apple', 'orange', 'banana', 'grapes', 'water melon']
values = [100, 200, 350, 150, 400]
plt.bar(fruits, values, color = 'blue', width = 0.5)
plt.title("Bar Graph")
plt.xlabel("category")
plt.ylabel("values")
plt.show()
```

```python
#Horizontal Bar chart
#bar chart
fruits = ['apple', 'orange', 'banana', 'grapes', 'water melon']
values = [100, 200, 350, 150, 400]
plt.barh(fruits, values)
plt.title("Bar Graph")
plt.xlabel("category")
plt.ylabel("values")
plt.show()
```

```python
#Mutiple Bar Graphs
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 

IT = [12, 30, 1, 8, 22] 
ECE = [28, 6, 16, 5, 10] 
CSE = [29, 3, 24, 25, 17] 

br1 = np.arange(len(IT)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

plt.bar(br1, IT, color ='r', width = barWidth, 
        edgecolor ='grey', label ='IT') 
plt.bar(br2, ECE, color ='g', width = barWidth, 
        edgecolor ='grey', label ='ECE') 
plt.bar(br3, CSE, color ='b', width = barWidth, 
        edgecolor ='grey', label ='CSE') 

plt.xlabel('Branch', fontweight ='bold', fontsize = 15) 
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15) 
plt.xticks([r + barWidth for r in range(len(IT))], 
        ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show() 
```

```python
#Stacked Bar chart
N = 5
boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(N)   
width = 0.35  

fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width,
             bottom = boys, yerr = girlStd)

plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))

plt.show()
```

#Scatter Plots in Matplotlib

catter plots are one of the most fundamental tools for visualizing relationships between two numerical 
variables.

```python
x = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y = np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])

plt.scatter(x, y)
plt.title("Basic Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

attributes in matplotlib - s, c, marker, cmap, alpha, edgecolors, label

```python
x1 = np.array([160, 165, 170, 175, 180, 185, 190, 195, 200, 205])
y1 = np.array([55, 58, 60, 62, 64, 66, 68, 70, 72, 74])

x2 = np.array([150, 155, 160, 165, 170, 175, 180, 195, 200, 205])
y2 = np.array([50, 52, 54, 56, 58, 64, 66, 68, 70, 72])

plt.scatter(x1, y1, color='blue', label='Group 1')
plt.scatter(x2, y2, color='red', label='Group 2')

plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Comparison of Height vs Weight between two groups')
plt.legend()
plt.show()
```

```python
x = np.array([3, 12, 9, 20, 5, 18, 22, 11, 27, 16])
y = np.array([95, 55, 63, 77, 89, 50, 41, 70, 58, 83])

a = [20, 50, 100, 200, 500, 1000, 60, 90, 150, 300] # size
b = ['red', 'green', 'blue', 'purple', 'orange', 'black', 'pink', 'brown', 'yellow', 'cyan'] # color

plt.scatter(x, y, s=a, c=b, alpha=0.6, edgecolors='w', linewidths=1)
plt.title("Scatter Plot with Varying Colors and Sizes")
plt.show()
```

#Histograms in Matplotlib

Histograms are one of the most fundamental tools in data visualization. They provide a graphical 
representation of data distribution, showing how frequently each value or range of values occurs

The X-axis represents intervals (called bins) of the data.
The Y-axis represents the frequency of values within each bin.

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram plot")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
# Generate random data for the histogram
data = np.random.randn(1000)

# Creating a customized histogram with a density plot
sns.histplot(data, bins=30, kde=True, color='lightgreen', edgecolor='red')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Customized Histogram with Density Plot')

# Display the plot
plt.show()
```

```python
data1 = np.random.randn(1000)
data2 = np.random.normal(loc=3, scale=1, size=1000)

# Creating a stacked histogram
plt.hist([data1, data2], bins=30, stacked=True, color=['cyan', 'Purple'], edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')

# Adding legend
plt.legend(['Dataset 1', 'Dataset 2'])

# Display the plot
plt.show()
```

#Pie Charts in Matplotlib

```python
A Pie Chart is a circular statistical plot that can display only one series of data. The area of the 
mchart is the total percentage of the given data. 
```

```python
costs = ['food', 'rent', 'bills', 'medicines', 'shopping', 'others']
values = [30, 15, 30, 10, 12, 3]
fig = plt.figure(figsize=(10, 7))
plt.pie(values, labels=costs)
plt.title("Pie Chart")
plt.show()
```

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]


# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)

# Creating color parameters
colors = ("orange", "cyan", "brown",
          "grey", "indigo", "beige")

# Wedge properties
wp = {'linewidth': 1, 'edgecolor': "green"}

# Creating autocpt arguments


def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)


# Creating plot
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct=lambda pct: func(pct, data),
                                  explode=explode,
                                  labels=cars,
                                  shadow=True,
                                  colors=colors,
                                  startangle=90,
                                  wedgeprops=wp,
                                  textprops=dict(color="magenta"))

# Adding legend
ax.legend(wedges, cars,
          title="Cars",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
ax.set_title("Customizing pie chart")

# show plot
plt.show()
```
