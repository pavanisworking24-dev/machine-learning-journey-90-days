#List Comprehension

List comprehension is a concise and powerful way to create new lists by applying an expression 
to each item in an existing iterable (like a list, tuple or range). It helps you write clean,
readable and efficient code compared to traditional loops.

```python
#example programs
lst = [1, 2, 3, 4, 5]
rse = [val*2 for val in lst]
print(rse)
```

```python
a = [1, 2, 3, 4, 5]
res = [val for val in a if val % 2 == 0]
print(res)
```

```python
c = [(x, y) for x in range(3) for y in range(3)]
print(c)
```

#Python Exception Handling

Python Exception Handling allows a program to gracefully handle unexpected events (like 
invalid input or missing files) without crashing

```python
try:
    n = 0
    res = 100 / n
    
except ZeroDivisionError:
    print("You can't divide by zero!")
    
except ValueError:
    print("Enter a valid number!")
    
else:
    print("Result is", res)
    
finally:
    print("Execution complete.")
```

#FILE HANDLING IN PYTHON

File handling refers to the process of performing operations on a file, such as 
creating, opening, reading, writing and closing it through a programming interface. I

```python
#opening a file
f = open("hello world.txt", "r")
print(f)
```

```python
#closing a file
f.close()
```

```python
#file properties
file = open("hello world.txt", "r")
print(file.name)
print(file.mode)
print(file.closed)
```

```python
file.close()
print("Is closed?", file.closed)
```

```python
#Reading a file
f = open("hello world.txt", "r")
content = f.read()
print(content)
f.close()
```

```python
with open("pavan.txt", "r") as f1:
    cont = f1.read(25)
    print(cont)
    f1.close()
```

```python
#writting a file
with open("pavan.txt", "w") as f:
    f.write("My name is Prabha")
    f.write("I am introvert")
print("file written sucessfully")
```

```python
with open("pavan.txt", "r") as p:
    content = p.read()
    print(content)
    p.close()
```
