#Variables
In Python, variables are used to store data that can be referenced and manipulated during
program execution. A variable is essentially a name that is assigned to a value.

```python
x = 25
s = "srikanth"
y = 4.3
print(x, y, s)
```

Python Operators

```python
"""
In Python programming, Operators in general are used to perform operations on values and variables.
Operators: Special symbols like -, + , * , /, etc.
Operands: Value on which the operator is applied.
"""
```

```python
#Types of operators in Python:
"""Arithmetic operators
Assignment operators
Comparison (Relational) operators
Logical operators
Bitwise operators
Membership operators
Identity operators"""
```

```python
#Arthematic Operators
# Variables
a = 15
b = 4
# Addition
print("Addition:", a + b)  
# Subtraction
print("Subtraction:", a - b) 
# Multiplication
print("Multiplication:", a * b)  
# Division
print("Division:", a / b) 
# Floor Division
print("Floor Division:", a // b)  
# Modulus
print("Modulus:", a % b) 
# Exponentiation
print("Exponentiation:", a ** b)
```

```python
#Relational Operators
a = 13
b = 33

print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

```python
#logical operators
= True
b = False
print(a and b)
print(a or b)
print(not a)
```

```python
#Bitwise operators
a = 10
b = 4

print(a & b)
print(a | b)
print(~a)
print(a ^ b)
print(a >> 2)
print(a << 2)
```

```python
#Assignment Operators
a = 10
b = a
print(b)
b += a
print(b)
b -= a
print(b)
b *= a
print(b)
b <<= a
print(b)
```

```python
#memebership operators
x = 24
y = 20
list = [10, 20, 30, 40, 50]
if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")

if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
```

```python
#Ternary operator
a, b = 10, 20
min = a if a < b else b
print(min)
```

Conditional Statements

```python
"""
if
if - else
if-elif-else
Nested if
"""
```

```python
#if statements
age = 20
if age >= 18:
    print("Eligible to vote.")
```

```python
#if - else
age = 10
if age <= 12:
    print("Travel for free.")
else:
    print("Pay for ticket.")
```

```python
#if-elif-else
age = 25

if age <= 12:
    print("Child.")
elif age <= 19:
    print("Teenager.")
elif age <= 35:
    print("Young adult.")
else:
    print("Adult.")
```

```python
#nested if 
age = 70
is_member = True
if age >= 60:
    if is_member:
        print("30% senior discount!")
    else:
        print("20% senior discount.")
else:
    print("Not eligible for a senior discount.")
```
