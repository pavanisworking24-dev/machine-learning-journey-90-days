#Pandas Merging, Joining and Concatenating

```python
#concatinating
data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']}

data2 = {'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'],
         'Age': [17, 14, 12, 52],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1, index=[0, 1, 2, 3])

df1 = pd.DataFrame(data2, index=[4, 5, 6, 7])

print(df, "\n\n", df1)
```

```python
frames = [df, df1]
print(pd.concat(frames))   # used when cloumns are same and rows are different or continous
```

```python
data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd'],
         'Mobile No': [97, 91, 58, 76]}

data2 = {'Name': ['Gaurav', 'Anuj', 'Dhiraj', 'Hitesh'],
         'Age': [22, 32, 12, 52],
         'Address': ['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
         'Qualification': ['MCA', 'Phd', 'Bcom', 'B.hons'],
         'Salary': [1000, 2000, 3000, 4000]}

df = pd.DataFrame(data1, index=[0, 1, 2, 3])

df1 = pd.DataFrame(data2, index=[2, 3, 6, 7])

print(df, "\n\n", df1)
```

```python
res = pd.concat([df, df1], axis = 1, join="inner")
print(res) # used when we want common rows in same datasets with all cloumns
```

```python
res2 = pd.concat([df, df1], axis=1, join="outer")
print(res2) #used when we want all the rows and all the cloumns
```

```python
res = pd.concat([df, df1], ignore_index=True)
res
```

```python
res1 = pd.concat([df, df1])
print(res1)
```

```python
#Concatenating Mixed DataFrames and Series
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}

df = pd.DataFrame(data1,index=[0, 1, 2, 3])

s1 = pd.Series([1000, 2000, 3000, 4000], name='Salary')

print(df, "\n\n", s1)
```

#merging

```python
merging is used when no columns are common in two datasets, only one cmmon column used to join
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1)

df1 = pd.DataFrame(data2)


print(df, "\n\n", df1)
```

```python
res = pd.merge(df, df1, on ="key")
print(res)
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1)

df1 = pd.DataFrame(data2)


print(df, "\n\n", df1)
```

```python
#merging dataFrames using multiple keys
res = pd.merge(df, df1, on=['key', 'key1'])
print(res)
```

```python
#Left join 
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1)

df1 = pd.DataFrame(data2)


print(df, "\n\n", df1)
```

```python
res = pd.merge(df, df1, how="left", on=['key', 'key1'])
print(res)
```

```python
#Right join
res = pd.merge(df, df1, how="right", on=['key', 'key1'])
print(res)
```

```python
#inner join
res = pd.merge(df, df1, how="inner", on=['key', 'key1'])
print(res)
```

```python
#outer join
res = pd.merge(df, df1, how="outer", on=['key', 'key1'])
print(res)
```

#Sorting in Pandas DataFrame

```python
#Sorting by single column
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 95, 80]}
df = pd.DataFrame(data)

sorted_df = df.sort_values(by='Age')
print(sorted_df)
```

```python
sort_df = df.sort_values(by="Age", ascending = False)
print(sort_df)
```

```python
#sorting by multiple columns
sort_df = df.sort_values(by=["Age", "Score"])
print(sort_df)
```

```python
data_with_nan = {"Name": ["Alice", "Bob", "Charlie", "David"],"Age": [28, 22, None, 22]}
df_nan = pd.DataFrame(data_with_nan)
print(df_nan)
```

```python
#sorting dataframe with nan
sort_df = df_nan.sort_values(by="Age", na_position = "first")
print(sort_df)
```

#Pivot Tables in Pandas

```python
df = pd.DataFrame({'Product': ['Carrots', 'Broccoli', 'Banana', 'Banana',
                               'Beans', 'Orange', 'Broccoli', 'Banana'],
                   'Category': ['Vegetable', 'Vegetable', 'Fruit', 'Fruit',
                                'Vegetable', 'Fruit', 'Vegetable', 'Fruit'],
                   'Quantity': [8, 5, 3, 4, 5, 9, 11, 8],
                   'Amount': [270, 239, 617, 384, 626, 610, 62, 90]})
df
```

```python
#Total sales product wise
res_piv = df.pivot_table(index=["Product"], values=["Amount"], aggfunc="sum")
print(res_piv)
```

```python
#Total sales category wse
res_piv = df.pivot_table(index=["Category"], values=["Amount"], aggfunc="sum")
print(res_piv)
```

```python
#total sales by both product and category wise
res_piv = df.pivot_table(index=["Product", "Category"], values=["Amount"], aggfunc="sum")
print(res_piv)
```

```python
#getting the mean, median , min product wise
res_piv = df.pivot_table(index = ["Product"], values = ["Amount"], aggfunc=["mean", "median", "min"])
print(res_piv)
```

```python
#getting the mean, median , min category wise
res_piv = df.pivot_table(index = ["Category"], values = ["Amount"], aggfunc=["mean", "median", "min"])
print(res_piv)
```
